```python
from llama_cpp import Llama

#2-bit quantized model
llm = Llama.from_pretrained(
	repo_id="TheBloke/OpenAssistant-Llama2-13B-Orca-8K-3319-GGUF",
	filename="openassistant-llama2-13b-orca-8k-3319.Q2_K.gguf",
)
```

```python
output = llm(
	"Once upon a time,",
	max_tokens=512,
	echo=True
)
print(output)
```

```python
output = llm(
	"Who is Narendra Modi?",
	max_tokens=100,
	echo=True
)
print(output)
```
